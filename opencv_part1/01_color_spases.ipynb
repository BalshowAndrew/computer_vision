# Цветовые пространства

## Загрузка изображения из файла

```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('datas/lena.png', cv2.IMREAD_COLOR)
img_gray = cv2.imread('datas/lena.png', cv2.IMREAD_GRAYSCALE)
```

```python
print('type:', type(img))
print('shape', img.shape)
```

```python
print("Высота:" + str(img.shape[0]))
print("Ширина:" + str(img.shape[1]))
print("Количество каналова:" + str(img.shape[2]))
```

## Визуализация в matplotlib

### Цветное изображение

```python
plt.imshow(img);
```

Библиотека OpenCV хранит канаты в порадке BGR, а не RGB.

Изменим порядок каналов для более корректной визуализации

```python
# Вариант 1
plt.imshow(img[...,::-1]);
```

```python
# Вариант 2
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));
```

### Доступ к значению пикселя

```python
(b, g, r) = img[1, 2]
print(f"Красный: {r}, Зеленый: {g}, Синий: {b}")
```

## Переход в пространство цветов HSV

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
img_hsv.shape
```

```python
plt.imshow(img_hsv);
```

```python
plt.figure(1, figsize = (16, 4))
plt.subplot(141)
plt.imshow(img[...,::-1])
plt.subplot(142)
plt.imshow(img_hsv[...,0])
plt.title('Hue')
plt.subplot(143)
plt.imshow(img_hsv[...,1])
plt.title('Saturation')
plt.subplot(144)
plt.imshow(img_hsv[...,2])
plt.title('Value');
```

```python
plt.figure(1, figsize = (16, 4))
plt.subplot(141)
plt.imshow(img[...,::-1])
plt.subplot(142)
plt.imshow(img_hsv[...,0], cmap='gray')
plt.title('Hue')
plt.subplot(143)
plt.imshow(img_hsv[...,1], cmap='gray')
plt.title('Saturation')
plt.subplot(144)
plt.imshow(img_hsv[...,2], cmap='gray')
plt.title('Value');
```
